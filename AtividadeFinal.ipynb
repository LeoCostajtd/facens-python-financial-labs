## Storage

```python
class Storage():
    def __init__(self, path):
        self.__path = path

    @property
    def path(self):
        return self.__path

    # Fun√ß√£o de leitura de dados
    def read_from_json(self):
        path = self.path
        if os.path.exists(path):
            with open(path, "r") as file:
                data_dict = json.load(file)
            return data_dict
        else:
            return False


    # Fun√ß√£o para salvar os dados
    def save_to_json(self, data_dict):
        path = self.path
        with open(path, "w") as file:
            json.dump(data_dict, file, indent=4)
```

```python
class System():

    def __init__(self, path):
        self.__options_menu = 'A - Hot√©is\nB - Reservas\nS - Sair'
        self.__options_menu_inputs = list(map(lambda op: op[0],self.__options_menu.split('\n')))

        self.__options_sub_menu = 'A - Cadastrar\nB - Visualizar\nC - Editar\nD - Excluir\nV - Voltar para o menu principal'
        self.__options_sub_menu_inputs = list(map(lambda op: op[0],self.__options_sub_menu.split('\n')))

        self.__options_return = 'V - Voltar para o menu principal\nOutra entrada - Voltar √† fun√ß√£o'

        self.__hotels_list = Hotels(path)


    @property
    def hotels_list(self):
        return self.__hotels_list


    @hotels_list.setter
    def hotels_list(self, hotels_list):
        self.__hotels_list = hotels_list


    # Fun√ß√£o que inicializa o sistema, realizando o processo de persistir informa√ß√µes em arquivo
    def start(self):
        self.show_menu()
        self.hotels_list.save_hotels()


    def show_menu(self):
        opt = ''
        while opt != 'S':
            opt = self.display_menu(self.__options_menu)
            if opt != 'S' and self.check_command(self.__options_menu_inputs, opt):
                self.show_sub_menu(opt)
        print('\nSistema finalizado!')
        return


    def show_sub_menu(self, opt):
        sub_opt = ''
        while sub_opt != 'V':
            print(f'{5*"="} Manipula√ß√£o de {"Hot√©is" if opt == "A" else "Reservas"} {5*"="}\n')
            sub_opt = self.display_menu(self.__options_sub_menu)
            if sub_opt != 'V' and self.check_command(self.__options_sub_menu_inputs, sub_opt):
                self.func_loop(opt, sub_opt)
        output.clear()
        return


    def display_menu(self, options_list):
        print(options_list)
        in_opt = input('\nInforme o comando desejado: ')
        return in_opt.upper()


    def check_command(self, options_list, opt):
        output.clear()
        if opt not in options_list:
            print('O comando informado √© inv√°lido!\n')
            return False
        else:
            return True


    def func_loop(self, opt, sub_opt):
        status = ''
        while status != 'V':
            status = self.run_command(opt, sub_opt)
            status = self.display_menu(self.__options_return) if status == 'F' else 'V'
            output.clear()
        return


    def run_command(self, opt, sub_opt):
        status = ''
        if opt == 'A':
            if sub_opt == 'A':
                status = self.create_hotel()

            elif sub_opt == 'B':
                status = self.list_hotels()

            elif sub_opt == 'C':
                status = self.update_hotel()

            elif sub_opt == 'D':
                status = self.delete_hotel()


        elif opt == 'B':
            if sub_opt == 'A':
                pass

            elif sub_opt == 'B':
                pass

            elif sub_opt == 'C':
                pass

            elif sub_opt == 'D':
                pass

        return status


    def input_hotel_info(self):
        print('Informe os dados do hotel\n')
        cnpj = input('CNPJ: ')
        name = input('Nome: ')
        location = input('Localiza√ß√£o: ')
        num_rooms_str = input('N√∫mero de quartos: ')

        if not (cnpj and name and location and num_rooms_str):
            print('\nTodos os campos s√£o obrigat√≥rios.\n')
            return []

        if num_rooms_str.isdigit():
            num_rooms = int(num_rooms_str)
            if not (num_rooms > 0):
                print('\nO n√∫mero de quartos deve ser um n√∫mero inteiro positivo.\n')
                return []
        else:
            print('\nO n√∫mero de quartos deve ser um valor num√©rico.\n')
            return []

        return [cnpj, name, location, num_rooms]


    def print_hotel_infos(self):
        if len(self.hotels_list.hotels) == 0:
            print('Nenhum hotel cadastrado!\n')
        print(f'{5*"="} Hot√©is cadastrados {5*"="}\n')
        for i, hotel in enumerate(self.hotels_list.hotels):
            print(f'\n{3*"="} Hotel {i+1}\n')
            hotel.show_infos()


    def select_hotel(self):
        number_str = input('Informe o numero do hotel desejado: ')

        if not number_str:
            print('\nO n√∫mero do hotel √© obrigat√≥rio.\n')
            return 'F'

        if number_str.isdigit():
            number = int(number_str)
            if not (1 <= number <= len(self.hotels_list.hotels)):
                print('\nInsira um n√∫mero de hotel v√°lido.\n')
                return 'F'
        else:
            print('\nInsira um n√∫mero de hotel v√°lido.\n')
            return 'F'

        return number


    def create_hotel(self):
        hotel_data = self.input_hotel_info()
        if len(hotel_data) == 0:
            return 'F'

        self.hotels_list.add_hotel_to_hotels(hotel_data)

        print(f'\nHotel cadastrado com sucesso!\n')
        self.wait()


    def list_hotels(self):
        self.print_hotel_infos()
        self.wait()


    def update_hotel(self):
        self.print_hotel_infos()
        index = self.select_hotel()

        if index == 'F':
            return 'F'
        else:
            output.clear()
            index -= 1

        hotel_data = self.input_hotel_info()
        if len(hotel_data) == 0:
            return 'F'

        self.hotels_list.update_hotel_in_hotels(hotel_data, index)

        print(f'\nDados atualizados com sucesso!\n')
        self.wait()


    def delete_hotel(self):
        self.print_hotel_infos()
        index = self.select_hotel()

        if index == 'F':
            return 'F'
        else:
            output.clear()
            index -= 1

        self.hotels_list.hotels[index].show_infos()
        self.hotels_list.delete_hotel_from_hotels(index)

        print(f'Hotel exclu√≠do com sucesso!\n')
        self.wait()


    def wait(self):
        input('Aperte qualquer tecla para voltar ao menu.')

```

## Hotel & Hotels

```python
class Hotel():

    def __init__(self, cnpj, name, location, num_rooms, reservations=[]):
        self.__cnpj = cnpj
        self.__name = name
        self.__location = location
        self.__total_rooms = num_rooms
        self.__reservations = reservations if len(reservations) else [True] * num_rooms

    # Getters and Setters
    @property
    def cnpj(self):
        return self.__cnpj

    @cnpj.setter
    def cnpj(self, cnpj):
        self.__cnpj = cnpj

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def location(self):
        return self.__location

    @location.setter
    def location(self, location):
        self.__location = location

    @property
    def total_rooms(self):
        return self.__total_rooms

    @total_rooms.setter
    def total_rooms(self, total_rooms):
        self.__total_rooms = total_rooms

    @property
    def reservations(self):
        return self.__reservations

    @reservations.setter
    def reservations(self, reservations):
        self.__reservations = reservations

    def reservate_room(self, room):
        self.__reservations[room] = not self.__reservations[room]

    # Methods
    def show_infos(self):
        print(f'CNPJ: {self.cnpj}')
        print(f'Nome: {self.name}')
        print(f'Localiza√ß√£o: {self.location}')
        print(f'Total de quartos: {self.total_rooms}')
        available_rooms  = [room for room, status in enumerate(self.reservations) if status == True]
        rooms = 'N√£o h√° quartos dispon√≠veis!'
        for i, room in enumerate(available_rooms):
            if i == 0:
                rooms = str(room+1).rjust(3)
            elif i%10 == 0:
                rooms += '\n' + str(room+1).rjust(3)
            else:
                rooms += ' - ' + str(room+1).rjust(3)
        print(f'Quartos dispon√≠veis: \n{rooms}\n')

    def change_num_rooms(self, num_rooms):
        current_num_rooms = len(self.__reservations)
        highest_reserved_room = 0

        for index, status in enumerate(self.__reservations):
            if not status:
                highest_reserved_room = max(highest_reserved_room, index + 1)

        if num_rooms >= highest_reserved_room:
            if num_rooms > current_num_rooms:
                self.__reservations.extend([True] * (num_rooms - current_num_rooms))
            elif num_rooms < current_num_rooms:
                self.__reservations = self.__reservations[:num_rooms]
            self.total_rooms = num_rooms
        else:
            print('N√£o √© poss√≠vel diminuir o n√∫mero de quartos pois h√° quartos reservados.')
            return False
        return True

    def update_hotel(self, hotel_data):
        cnpj, name, location, num_rooms = hotel_data
        if self.change_num_rooms(num_rooms):
            self.cnpj = cnpj
            self.name = name
            self.location = location
            self.num_rooms = num_rooms

    def make_reservation(self, room):
        if 1 <= room <= len(self.reservations):
            if self.reservations[room - 1]:
                self.reservate_room(room - 1)
                print(f'Reserva do quarto {room} efetuada com sucesso.')
            else:
                print(f'O quarto {room} j√° est√° ocupado.')
        else:
            print('Quarto inv√°lido.')

    def cancel_reservation(self, room):
        if 1 <= room <= len(self.reservations):
            if not self.reservations[room - 1]:
                self.reservate_room(room - 1)
                print(f'Reserva do quarto {room} canceled.')
            else:
                print(f'O quarto {room} j√° est√° dispon√≠vel.')
        else:
            print('Quarto inv√°lido.')



```

```python
class Hotels():

    def __init__(self, path):
        self.__path = path
        self.__hotels = self.load_hotels()

    @property
    def path(self):
        return self.__path

    @path.setter
    def path(self, path):
        self.__path = path

    @property
    def hotels(self):
        return self.__hotels

    @hotels.setter
    def hotels(self, hotels):
        self.__hotels = hotels

    def load_hotels(self):
        hotels_dict = Storage(self.path).read_from_json()
        if hotels_dict != False:
            return [ Hotel(*[hotel[label] for label in hotel]) for hotel in hotels_dict ]
        else:
            return []

    def save_hotels(self):
        data_dict = [hotel.__dict__ for hotel in self.hotels]
        Storage(self.path).save_to_json(data_dict)

    def add_hotel_to_hotels(self, hotel_data):
        self.__hotels.append(Hotel(*hotel_data))

    def update_hotel_in_hotels(self, hotel_data, index):
        self.__hotels[index].update_hotel(hotel_data)

    def delete_hotel_from_hotels(self, index):
        self.__hotels.pop(index)
```

```python
import re

def validar_cnpj(cnpj):
    # Remover caracteres n√£o num√©ricos
    cnpj = re.sub(r'[^0-9]', '', cnpj)

    # Verificar se o CNPJ possui 14 d√≠gitos
    if len(cnpj) != 14:
        return False

    # Verificar se todos os d√≠gitos s√£o iguais (CNPJs inv√°lidos)
    if len(set(cnpj)) == 1:
        return False

    return True

# Teste
cnpj = "00.000.000/0001-91"
if validar_cnpj(cnpj):
    print("CNPJ v√°lido")
else:
    print("CNPJ inv√°lido")
```
